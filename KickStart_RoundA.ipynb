# K - Goodness String

```python
def goodness_score(s, k):
  min = 0
  x = 0
  for i in range(0, (len(s)//2)):
    if s[i] != s[len(s) - i - 1]:
      x += 1
  
  if x == k:
    return 0
  elif x > k:
    minOperations = x - k
  else:
    minOperations = k - x

  return minOperations

goodness_score("CABABC", 4)
```

```python
t = int(input())
case = []
for i in range(t):
  v = map(lambda x: int(x),input().split(" "))
  s = input()
  case.append([s, list(v)[1]])

for i in range(t):
  score = goodness_score(case[i][0], case[i][1])
  print(f"Case #{i}: {score}")
```
