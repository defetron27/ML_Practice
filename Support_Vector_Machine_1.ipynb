```python
import pandas as pd

df = pd.read_csv("/content/gdrive/My Drive/datasets/cricket_players_without_index.csv")

df.head()
```

```python
df.loc[6:9, ['player_name', 'runs']]
```

```python
df1 = df.loc[:, ['batting_avg', 'bowling_avg', 'wickets', 'runs']]

df1
```

```python
df1['target_name'] = df.loc[:, ['role']]

df1
```

```python
from sklearn import preprocessing

label_encoding = preprocessing.LabelEncoder()

df1['target'] = label_encoding.fit_transform(df1['target_name'])

df1
```

```python
X = df1.loc[:, 'batting_avg':'runs']

y = df1.loc[:, 'target']
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

X_train.head()
```

```python
from sklearn.svm import SVC

model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

y_pred
```

```python
import numpy as np
import matplotlib.pyplot as plt

np_test = np.arange(len(y_test)).reshape((len(y_test), 1))

plt.scatter(np_test, y_test, color='red', marker='.')
plt.plot(np_test, y_pred, color='blue')
```

```python
model.fit(X, y)

y_pred = model.predict(X)

np_test = np.arange(len(y)).reshape((len(y), 1))

plt.scatter(np_test, y, color='red', marker='.')
plt.plot(np_test, y_pred, color='blue')
```

```python
import matplotlib.colors as colors

markers = ['*', '.', '+', ',']

cmap = plt.cm.rainbow
norm = colors.Normalize(vmin=y.min(), vmax=y.max())

for i in range(4):
    yy = X[y[:] == i]['runs']
    XX = np.arange(len(yy)).reshape((len(yy), 1))
    
    plt.scatter(XX, yy, color=cmap(norm(i)), marker=markers[i])
```
