```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X, Y)
```

```python
print(model.intercept_)
```

```python
coeff_parameter = pd.DataFrame(model.coef_,X.columns,columns=['Coefficient'])

coeff_parameter
```

```python
y_pred = model.predict(X)

y_pred
```
