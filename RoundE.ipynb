```python
def swap(ch, i, j):
    temp = ch[i]
    ch[i] = ch[j]
    ch[j] = temp
 

def permutations(ch, curr_index=0):
 
    if curr_index == len(ch) - 1:
        print(''.join(ch))
 
    for i in range(curr_index, len(ch)):
        swap(ch, curr_index, i)
        permutations(ch, curr_index + 1)
        swap(ch, curr_index, i)
```

```python
ss_lst = []

def swap(ch, i, j):
    temp = ch[i]
    ch[i] = ch[j]
    ch[j] = temp
 

def permutations(s, ch, curr_index=0):
 
    if curr_index == len(ch) - 1:
        ss = "".join(ch)
        ss_len = len(ss)
        c = 0
        for i in range(ss_len):
          if s[i] == ss[i]:
            break;
          else:
            c += 1
        
        if c == ss_len:
          print(c, ss)
          ss_lst.append(ss)
          return
 
    for i in range(curr_index, len(ch)):
        swap(ch, curr_index, i)
        permutations(s, ch, curr_index + 1)
        swap(ch, curr_index, i)
```

```python
s = "start"
permutations(s, list(s))
```
