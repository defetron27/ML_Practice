# Neural Network Implementation from Scratch

```
# We are going to do is implement the “OR” logic gate using a perceptron.
# Keep in mind that here we are not going to use any of the hidden layers.
```

What is logical OR Gate?

```
# - Straightforwardly, when one of the inputs is 1, the output of the OR gate is going to be 1.
# - It means that the output is 0 only when both of the inputs are 0.
```

What is a Perceptron?

```
# - A perceptron is a neural network without any hidden layer.
# - A perceptron only has an input layer and an output layer.
```

```
# implementing the OR gate
input = [0, 1]

target = 1

weight = [0.2, 0.3]

bias = 0.5

o1 = bias * target

for i, j in zip(input, weight):
  o1 = o1 + (i * j)

o1
```

```
# calculate the value
# f(X) = 1 / 1 + e^-X
# output of 01(o1)

# don't do this -> 1 / 1 + np.exp(-o1)

output = 1 / (1 + np.exp(-o1))

output
```

Error Calculation:

```
# Next, we are going to use Mean Squared Error for calculating the error :

# MSE -> mean squared error

# MSE = E 1/2 * (target - output)^2

# The summation sign (Sigma symbol) means that we have to add our error for all our input sets.
# Here we are going to see how that works for only one input set.

mse = 1/2 * ((target - output) ** 2)

mse
```

```
# We have to do the same for all the remaining inputs.
# Now that we have found the error, we have to update the values of weight to make the error minimum.
# For updating weight values, we are going to use a gradient descent algorithm.

# So, we are trying to get the value of weight such that the error becomes minimum.
# We need to figure out whether we need to increase or decrease the weight value.
# Once we know that, we keep on updating the weight value in that direction until error becomes minimum.
# We might reach a point where if further updates occur to the weight, the error will increase.
# At that time, we need to stop, and that is our final weight value.
# In real-life data, the situation can be a bit more complex.
# In the example above, we saw that we could try different weight values and get the minimum error manually.
# However, in real-life data, weight values are often decimal (non-integer).
# Therefore, we are going to use a gradient descent algorithm with a low learning rate so
#    that we can try different weight values and obtain the best predictions from our model.
```

What is Gradient Descent?

```
#   - Gradient Descent is a machine learning algorithm that operates iteratively to find the optimal values for its parameters.
#   - It takes into account, user-defined learning rate, and initial parameter values.

# Working: (Iterative)
#   1. Start with initial values.
#   2. Calculate cost.
#   3. Update values using the update function.
#   4. Returns minimized cost for our cost function

# Why do we need it?
#   - Generally, what we do is, we find the formula that gives us the optimal values for our parameter.
#   - However, in this algorithm, it finds the value by itself!

# X = X - lr * (d/dX f(X))
# where,
#   - X = input
#   - F(X) = output based on X
#   - lr = learning rate

# We are going to update our weight with this algorithm. First of all, we need to find the derivative f(X).

# Derivation of the formula used in a neural network:
#   - Next, what we want to find is how a particular weight value affects the error. To find that we are going to apply the chain rule.
#   - dError / dw = (dError / dOuto) * (dOuto / dIno) * (dIno / dw)

# Afterward, what we have to do is we have to find values for these three derivatives.

# Calculating derivatives:
# dError / dOuto = d / dOuto (1/2 * (target - output)^2)
# dError / dOuto = (1/2 * 2 * (target - output)) * (d / dOuto (target - output))
# dError / dOuto = (target - output) * (-1)
# dError / dOuto = output - target

# In our case:
# output = 0.68..
# target = 1

# dError / dOuto1 = (0.68.. - 1)

error = output - target

error
```

```
# Finding the second part of the derivative:
# dOuto / dIno = outo(1 - outo)

# Value of outo1:
# out1 = (1 / 1 + e^-ino1)

# Finding the derivative with respect to ino1:
# dOuto1 / dIno1 = d / dIno1 (1 / 1 + e^-ino1)

# Simplifying it a bit to find the derivative easily:
# dOuto1 / dIno1 = d / dIno1 (1 + e^-ino1)^-1

# Applying chain rule and power rule:
# dOuto1 / dIno1 = -1(1 + e^-ino1)^-2 * d / dIno1 (1 + e^-ino1)

# Applying sum rule:
# dOuto1 / dIno1 = -1(1 + e^-ino1)^-2 * (d / dIno1 * (1) + d / dIno1 * e^-ino1)

# The derivative of constant is zero:
# dOuto1 / dIno1 = -1(1 + e^-ino1)^-2 * (0 + d / dino1 (e^-ino1))

# Applying exponential rule and chain rule:
# dOuto1 / dIno1 = -1(1 + e^-ino1)^-2 * (e^-ino1 * d/dino1 [-ino1])

# Simplifying it a bit:
# dOuto1 / dIno1 = -1(1 + e^-ino1)^-2 * (e^-ino1 * -1)

# Multiplying both negative signs:
# dOuto1 / dIno1 = (1 + e^-ino1)^-2 * (e^-ino1)

# Put the negative power in the denominator:
# dOuto1 / dIno1 = (e^-ino1) + (1 + e^-ino1)^2

# That is it. However, we need to simplify it as it is a little complex for our machine learning algorithm to process for a large number of inputs.

# Simplifying it:
# dOuto1 / dIno1 = 1 * (e^-ino1) / (1 * e^-ino1) * (1 * e^-ino1)

# Further simplification:
# dOuto1 / dIno1 = 1/(e^-ino1) * (e^-ino1) / (1 + e^-ino1)

# Adding +1–1:
# dOuto1 / dIno1 = 1/(e^-ino1) * (e^-ino1) + 1 - 1 / (1 + e^-ino1)

# Separate the parts:
# dOuto1 / dIno1 = 1/(1+e^-ino1) * ((1 + e^-ino1) / (1 + e^-ino1) - 1/(1 + e^-ino1))

# Simplify:
# dOuto1 / dIno1 = 1/(1+e^-ino1) * (1 - 1/(1+e^-ino1))

# Now we all know the value of outo1 from equation 1:
# outo1 = (1 / 1 + e^-ino1)

# From that we can derive the following final derivative:
# dOuto1 / dIno1 = outo1 * (1 - outo1)

# Calculating the value of our input:
# dOuto1 / dIno1 = (0.68..) * (1 - 0.68..)\

oo1 = output * (1 - output)

oo1
```

```
# Finding the third part of the derivative :
# dIno / dw = input values

# Value of ino:
# ino1 = w1 * i1 + w2 * i2 + w3 * 1

# Finding derivative:
# All the other values except w2 will be considered constant here.
# dIno1 / dw2 = 0 + i2 + 0 = i2

# Calculating both values for our input:
# dIno1 / dw2 = i2 = 1

io1 = 1

# Putting it all together:
# (dError / dOuto) * (dOuto / dIno) * (dIno / dw) = ?

d = error * oo1 * io1

d
```

```
# Putting it in our main equation:
# w = w - lr * (dError / dw2)

w2 = 1 - (-.05) * (-d)

w2

# Notice that the value of the weight has increased here
```

```
# here,
# ino -> the value of input * weight
# outo -> value of output after applying sigmoid function

# a. Input for perceptron:
#   - ino = w1 * i1 + w2 * i2 + w3 * b

# b. Applying sigmoid function for predicted output:
#   - outo = 1 / (1 + e^-ino)

# c. Calculate the error:
#   - error = E 1/2 (target output - outo)

# d. Changing the weight value based on gradient descent formula:
#   - w = w - lr * (dError / dw)

# e. Calculating the derivative:
#   - dError / dw = (dError / dOuto) * (dOuto / dIno) * (dIno / dw)

# f. Individual derivatives:
#   - dError / dOuto = (outo - target output)

# dOuto / dIno = outo(1 - outo)
# dIno / dw = input values
```

```
input_features = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])

input_features.shape
```

```
target_output = np.array([0, 1, 1, 1])

target_output = target_output.reshape(4, 1)

target_output
```

```
# At this point, we are taking these values randomly.
# Here we have two input features, so we are going to take two weight values.

weights = np.array([[0.1], [0.2]])

weights.shape
```

```
# Adding Bias Values and Assigning a Learning Rate:
# bias = 1
# However, the weight assigned to it is random at first, and our model will optimize it for our target output.
# The other parameter is called the learning rate(LR).
# We are going to use the learning rate in a gradient descent algorithm to update the weight values.
# Generally, we keep the learning rate as low as possible so that we can achieve a minimum error rate.

bias = 0.3
lr = 0.05
```

```
# define the sigmoid function
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

```
# Derivative of sigmoid function:
# In gradient descent algorithm we are going to need the derivative of the sigmoid function.
def sigmoid_der(x):
  return (sigmoid(x) * (1 - sigmoid(x)))
```

```
# The main logic for predicting output and updating the weight values:
# main login for neural network:
# running our code 10000 times:

for epoch in range(10000):
  inputs = input_features

  # Feedforward input:
  in_o = np.dot(inputs, weights) + bias

  # Feedforward output:
  out_o = sigmoid(in_o)

  # Backpropogation
  # Calculating error
  error = out_o - target_output

  # Going with the formula:
  x = error.sum()
  print(x)

  # Calculating derivative:
  derror_douto = error
  douto_dino = sigmoid_der(out_o)

  # Multiplying individual derivatives:
  deriv = derror_douto * douto_dino

  # Multiplying with the 3rd individual derivative:
  # Finding the transpose of input_features:
  input = input_features.T
  deriv_final = np.dot(input, deriv)

  # Updating the weights values:
  weights -= lr * deriv_final

  # Updating the bias weight value:
  for i in deriv:
    bias -= lr * i
```

```
# Why do we add bias?
# Suppose if we have input values (0,0), the sum of the products of the input nodes and weights is always going to be zero.
# In this case, the output will always be zero, no matter how much we train our model.
# To resolve this issue and make reliable predictions, we use the bias term.
# In short, we can say that the bias term is necessary to make a robust neural network.
# Therefore, how does the value of bias affects the shape of our sigmoid function? Let’s visualize it with some examples.
# To change the steepness of the sigmoid curve, we can adjust the weight accordingly.
```

```
input = np.linspace(-10, 10, 100)

# output values:
output = sigmoid(input)
plt.plot(input, output, c="b")

# this will decrease the steepness:
output = sigmoid(.5 * input)
plt.plot(input, output, c="r")

# this will increase the steepness:
output = sigmoid(2.5 * input)
plt.plot(input, output, c="g")
```

```
# From the output, we can quickly notice that for negative values, the output of the sigmoid function is going to be ≤0.5.
# Moreover, for positive values, the output is going to be >0.5.
# From the figure (red curve), you can see that if we decrease the value of the weight,
#   it decreases the value of steepness, and if we increase the value of weight (green curve), it increases the value of steepness.
# However, for all of the three curves, if the input is negative, the output is always going to be ≤0.5.
# For positive numbers, the output is always going to be >0.5.
```

```
# What if we want to change this pattern?
# For such case scenarios, we use bias values.

# output values:
output = sigmoid(input)
plt.plot(input, output, c="b")

# this will decrease the steepness:
output = sigmoid(input + 5)
plt.plot(input, output, c="r")

# this will increase the steepness:
output = sigmoid(input - 5)
plt.plot(input, output, c="g")
```
